```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import data_provider
%matplotlib inline
dp = data_provider.DataProvider()
dp.load_data()
```

```python
x,y0 = dp.next_batch()
x,y1 = dp.get_test()
y = np.concatenate([y0[:,0],y1[:,0]])
plt.plot(list(range(y.size)),y,label='PM2.5')
plt.xlabel('time(day)')
plt.ylabel('PM2.5(μg/m3)')
plt.legend()
# plt.title('2014-01-01 至 2015-12-30 门头沟站点PM2.5 分布')
plt.show()
```

```python
rng = pd.date_range('1/1/2014 00:00', periods=y.size, freq='D')
target = pd.Series(y,rng)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
def testStationarity(ts):
    dftest = adfuller(ts)
    # 对上述函数求得的值进行语义描述
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput
print(testStationarity(target))
x = list(range(target.size))
# plt.plot(x,target)
f = plt.figure(facecolor='white')
ax1 = f.add_subplot(211)
plot_acf(target, lags=31, ax=ax1)

# ax2 = f.add_subplot(212)
# plot_pacf(target, lags=31, ax=ax2)
plt.show()
```

```python
def testStationarity(ts):
    dftest = adfuller(ts)
    # 对上述函数求得的值进行语义描述
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput

from statsmodels.tsa.arima_model import ARMA
result = []
for start in range(0,29):
    print(start)
    temp_target = target[start:start+700]
    print(testStationarity(temp_target))
    model = ARMA(temp_target, order=(3, 1)) 
    
    result_arma = model.fit( )
    predict_target = result_arma.predict()
    real_target = temp_target[predict_target.index]
    a,_,_ = result_arma.forecast(steps=1)
    result.append(a[0])
    
print(result)
    
```

```python
real = target[700:]
print(real.shape,len(result))
x = list(range(len(result)))
plt.plot(x,real,label='real')
plt.plot(x,result,label='arma')
plt.xlabel('test_day(day)')
plt.ylabel('PM2.5(μg/m3)')
plt.legend()
plt.show()
```
