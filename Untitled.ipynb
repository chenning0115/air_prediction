```python
def testStationarity(ts):
    dftest = adfuller(ts)
    # 对上述函数求得的值进行语义描述
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    return dfoutput
print(testStationarity(target_s))
x = list(range(target_s.size))
plt.plot(x,target_s)
f = plt.figure(facecolor='white')
ax1 = f.add_subplot(211)
plot_acf(target_s, lags=31, ax=ax1)
ax2 = f.add_subplot(212)
plot_pacf(target_s, lags=31, ax=ax2)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
rng = pd.date_range('1/1/2014 00:00', periods=target_s.size, freq='D')
df_target = pd.Series(target_s,rng)
decomposition = seasonal_decompose(df_target, model="additive")

trend = decomposition.trend
plt.plot(trend)
plt.show()
seasonal = decomposition.seasonal
plt.plot(seasonal)
plt.show()
residual = decomposition.resid
plt.plot(residual)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARMA
model = ARMA(df_target, order=(3, 3)) 
result_arma = model.fit( disp=-1, method='css')
predict_target = result_arma.predict()
real_target = df_target[predict_target.index]
x = list(range(real_target.size))
plt.plot(x,real_target)
plt.plot(x,predict_target)
plt.show()
```
